### Task 4: interval
This problem was asked by Snapchat.

Given a list of possibly overlapping intervals, return a new list of intervals where all overlapping intervals have been merged.

The input list is not necessarily ordered in any way.

For example, given [(1, 3), (5, 8), (4, 10), (20, 25)], you should return [(1, 3), (4, 10), (20, 25)].


Note cho moị người là bài tập optional thêm cho assignment là bài bài 4 (interval trong Python), bài 5 trong advanced python, và phần GroupBy của SQL

```
# Your code here
def overlap(arr):
  arr.sort(key = lambda d : d[0])
  tmp =[]
  start = arr[0][0]
  end = arr[0][1]
  for i in arr:
    if (i[0]<end):
      end = i[1]
    else:
      tmp.append((start, end))
      start = i[0]
      end = i[1]
  tmp.append((start, end))
  return tmp
a =  [(1, 3), (4, 5), (7, 9), (8, 10)]
print(overlap(a))
```

```
import sqlite3
conn = sqlite3.connect('chinook.db')
import pandas as pd
data = pd.read_sql_query('SELECT * FROM albums;', conn)
data.head()
```

```
# Find the albums of Aerosmith
pd.read_sql_query('''
SELECT * 
FROM albums 
INNER JOIN artists 
  ON  albums.ArtistId = artists.ArtistId
WHERE artists.Name = "Aerosmith"
''', conn)
```

```
# List first 10 tracks and its album's title order by track's name
pd.read_sql_query('''
SELECT * 
FROM tracks
INNER JOIN albums
    ON tracks.AlbumId = albums.AlbumId
ORDER BY tracks.Name ASC
LIMIT 10
''', conn)
```

```
# List all employees and who they report to
pd.read_sql_query('''
SELECT e.FirstName, e.Title, r.FirstName, r.Title
FROM employees e
INNER JOIN employees r
    ON r.EmployeeId = e.ReportsTo
''', conn)
```

```
# Find all tracks that couldn't be sold (null invoice_items)
pd.read_sql_query('''
SELECT * 
FROM tracks 
LEFT JOIN invoice_items 
     ON tracks.TrackId = invoice_items.TrackId
WHERE invoice_items.InvoiceId IS NULL
''', conn)
```

```
# List all genres and the distinct media types in each genre
pd.read_sql_query('''
SELECT DISTINCT genres.GenreId, genres.Name, media_types.Name 
FROM genres
INNER JOIN tracks
    ON genres.GenreId = tracks.GenreId

INNER JOIN media_types
    ON tracks.MediaTypeId = media_types.MediaTypeId
''', conn)
```

```
# Find the revenue of 2013
pd.read_sql_query('''
SELECT SUM(Total)
FROM invoices
WHERE InvoiceDate LIKE "2013%"
''', conn)
```

```
# Find the number of tracks belong to the genre Rock
pd.read_sql_query('''
SELECT COUNT(TrackId) 
FROM tracks
INNER JOIN genres
    ON tracks.GenreId = genres.GenreId
WHERE genres.Name = 'Rock'
''', conn)
```

```
# Find the number of customers each employee has supported
pd.read_sql_query('''
SELECT COUNT(SupportRepId), employeeId
FROM customers
INNER JOIN employees
    ON customers.SupportRepId = employees.EmployeeId
GROUP BY employeeId
''', conn)
```

```
# Find the customer who paid the most
pd.read_sql_query('''
SELECT CustomerId, FirstName, LastName
FROM(
SELECT COUNT(InvoiceId) as count, CustomerId as id
FROM invoices
GROUP BY CustomerId
)
INNER JOIN customers
    ON customers.CustomerId = id
WHERE (count = (SELECT MAX(count) FROM (SELECT COUNT(InvoiceId) as count, CustomerId
FROM invoices
GROUP BY CustomerId)))
''', conn)
```

```
# Find the richest artist
pd.read_sql_query('''
SELECT MAX(sum), ArtistId, Name
FROM(
SELECT art.ArtistId, art.Name, SUM(i.Total) as sum
FROM artists art
LEFT JOIN albums alb
    ON art.ArtistId = alb.ArtistId
LEFT JOIN tracks t
    ON alb.AlbumId = t.AlbumId
LEFT JOIN invoice_items it
    ON t.TrackId = it.TrackId
LEFT JOIN invoices i
    ON it.InvoiceId = i.InvoiceId
GROUP BY art.ArtistId
)
''', conn)
```
